```python
#Tocknize - Text Handling - Prepare
!pip install nltk
!pip install emoji
!pip install contractions
import re
import nltk
import emoji
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from contractions import fix
from nltk.corpus import wordnet
from nltk.tokenize.casual import TweetTokenizer
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('omw-1.4')
stop_words = set(stopwords.words('english'))
tokenizer = TweetTokenizer()
```

```python
#Clean and tockenize function
def clean_and_tokenize(text):

    # Expanding contractions (e.g., "don't" → "do not")
    text = fix(text)
    # Lowercasing
    text = text.lower()
    # Removing multiple spaces
    text = re.sub(r'\s+', ' ', text)
    # Removing punctuation
    text = re.sub(r'[^\w\s]', '', text)
    # Removing emojis
    text = emoji.replace_emoji(text, replace='')
    # Removing URLs
    text = re.sub(r'http\S+|www.\S+', '', text)
    # Removing email addresses
    text = re.sub(r'\S+@\S+', '', text)
    # Removing stop words
    text = ' '.join([word for word in text.split() if word not in stop_words])
    # Removing extra short words (e.g., "a", "an", "it")
    text = ' '.join([word for word in text.split() if len(word) > 1])

    words = tokenizer.tokenize(text)  # Tokenization
    return words

#Lemmatizer Function
from nltk import pos_tag
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer

def get_wordnet_pos(word):
    tag = pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ, "N": wordnet.NOUN, "V": wordnet.VERB, "R": wordnet.ADV}
    return tag_dict.get(tag, wordnet.VERB)

def lemmatize_tokens(tokens):
    lemmatizer = WordNetLemmatizer()
    return [lemmatizer.lemmatize(token, get_wordnet_pos(token)) for token in tokens]

#Stemming Function
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

def stem_tokens(tokens):
    return [stemmer.stem(word) for word in tokens]
```

```python
#Title and Text Cleaned and Tocknized
df_review["tokenized_review_title"] = df_review["review_title"].apply(clean_and_tokenize)
df_review["tokenized_review_text"] = df_review["review_text"].apply(clean_and_tokenize)
```

```python
#Title and Text Lemmantize
df_review["tokenized_review_title_lem"] = df_review["tokenized_review_title"].apply(lemmatize_tokens)
df_review["tokenized_review_text_lem"] = df_review["tokenized_review_text"].apply(lemmatize_tokens)
df_review.head()
```

```python
#Title and Text Stemming
df_review["tokenized_review_title_stem"] = df_review["tokenized_review_title"].apply(stem_tokens)
df_review["tokenized_review_text_stem"] = df_review["tokenized_review_text"].apply(stem_tokens)
df_review.head()
```

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from yellowbrick.cluster import KElbowVisualizer, SilhouetteVisualizer


sc = StandardScaler()

normalized_data = sc.fit_transform(embeddings)
df_normalized = pd.DataFrame(normalized_data)


kmeans_model = KMeans(init="k-means++", random_state=99)


visualizer = KElbowVisualizer(kmeans_model, k=(2, 14), metric="silhouette", timings=False)

visualizer.fit(df_normalized)
visualizer.show()
```

```python
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.metrics import silhouette_score

methods = ['single', 'complete', 'average', 'ward']

for method in methods:
    results=[]

    Z = linkage(df_normalized, method=method)

    for num_cluster in range(2, 10):
      labels = fcluster(Z, num_cluster, criterion='maxclust')
      score = silhouette_score(df_normalized, labels)
      results.append((num_cluster, score))

    df_results= pd.DataFrame(results, columns=['num_cluster', 'silhouette_score'])
    plt.plot(df_results['num_cluster'], df_results['silhouette_score'], marker='o', label=method)

plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Comparison of Linkage Methods using Silhouette Score')
plt.legend(title="Linkage Method")
plt.grid(True)
plt.show()
```

```python

n_clusters = 5

Z = linkage(df_normalized, 'average')
df_normalized['cluster_labels_average'] = fcluster(Z, n_clusters, criterion='maxclust')

plt.figure(figsize=(10, 5))
dendrogram(Z, p=50, truncate_mode='lastp')
plt.title("Method: Average Linkage")
plt.show()





Z = linkage(df_normalized, 'complete')
df_normalized['cluster_labels_complete'] = fcluster(Z, n_clusters, criterion='maxclust')

plt.figure(figsize=(10, 5))
dendrogram(Z, p=50, truncate_mode='lastp')
plt.title("Method: Complete Linkage")
plt.show()





Z = linkage(df_normalized, 'single')
df_normalized['cluster_labels_single'] = fcluster(Z, n_clusters, criterion='maxclust')

plt.figure(figsize=(10, 5))
dendrogram(Z, p=50, truncate_mode='lastp')
plt.title("Method: Single Linkage")
plt.show()




Z = linkage(df_normalized, 'ward')
df_normalized['cluster_labels_ward'] = fcluster(Z, n_clusters, criterion='maxclust')

plt.figure(figsize=(10, 5))
dendrogram(Z, p=50, truncate_mode='lastp')
plt.title("Method: Ward Linkage")
plt.show()






```
