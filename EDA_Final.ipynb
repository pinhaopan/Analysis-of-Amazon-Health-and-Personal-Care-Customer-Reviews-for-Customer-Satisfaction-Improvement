## Data Loading

```python
import gdown
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


url = 'https://drive.google.com/uc?export=download&id=1hVsTyC7DJH8pplo1DmVm13v8QDkONjwo'
output = 'review_list.json'
gdown.download(url, output, quiet=False)

df_review = pd.read_json(output, lines=True)
df_review = df_review.drop(columns=['images'])
df_review.rename(columns={'title': 'review_title'}, inplace=True)
df_review.rename(columns={'text': 'review_text'}, inplace=True)

df_review
```

```python
url = 'https://drive.google.com/uc?export=download&id=14dVOPZM7-hCEdXJswxlb_c6XDErC7lac'
output = 'meta_list.json'
gdown.download(url, output, quiet=False)

df_meta = pd.read_json(output, lines=True)
df_meta = df_meta.drop(columns=['main_category','features','description','price','images','videos','categories','bought_together'])
df_meta.rename(columns={'title': 'meta_title'}, inplace=True)

df_meta
```

## EDA

```python
rating_counts = df_review['rating'].value_counts(normalize=True).sort_index(ascending=False) * 100
ax = rating_counts.plot(kind='bar')
plt.xlabel('Rating')
plt.ylabel('Percentage')
plt.title('Distribution of Ratings')
plt.xticks(rotation=0)
ax.bar_label(ax.containers[0], fmt='%.1f%%')
plt.show()
```

```python
df_meta["brand"] = df_meta["details"].apply(lambda x: x.get("Brand") if isinstance(x, dict) else None)

brand_ratings = df_meta.groupby("brand")["rating_number"].sum().reset_index()
top_10_rating_number_brands = brand_ratings.sort_values("rating_number", ascending=False).head(10)

print(top_10_rating_number_brands)

plt.figure(figsize=(10, 6))
plt.bar(top_10_rating_number_brands["brand"], top_10_rating_number_brands["rating_number"], color="skyblue")
plt.xlabel("Brand")
plt.ylabel("Total Rating Number")
plt.title("Top 10 Brand - Accumulated Rating Number")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
store_ratings = df_meta.groupby("store")["rating_number"].sum().reset_index()

top_10_stores = store_ratings.sort_values("rating_number", ascending=False).head(10)
print(top_10_stores)

plt.figure(figsize=(10, 6))
plt.bar(top_10_stores["store"], top_10_stores["rating_number"], color="skyblue")
plt.xlabel("Store")
plt.ylabel("Accumulated Rating Number")
plt.title("Top 10 Stores - Accumulated Rating Number")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df_review['datetime'] = pd.to_datetime(df_review['timestamp'], unit='ms')

plt.figure(figsize=(8, 6))
plt.hist(df_review['datetime'], bins=100, edgecolor='black')
plt.xlabel('Date', fontsize=16)
plt.ylabel('Number of Reviews', fontsize=16)
plt.title('The Time Distribution of Users Review' ,fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.tight_layout()
```

```python
df_review['review_text_word_count'] = df_review['review_text'].apply(lambda x: len(x.split()) if isinstance(x, str) else 0)
df_review['review_text_word_count_cutoff'] = df_review['review_text_word_count'].apply(lambda x: x if x <= 300 else 300)
df_review['review_title_word_count'] = df_review['review_title'].apply(lambda x: len(x.split()) if isinstance(x, str) else 0)

df_meta['meta_title_word_count'] = df_meta['meta_title'].apply(lambda x: len(x.split()) if isinstance(x, str) else 0)
df_meta['average_rating_threshold'] = np.ceil(df_meta['average_rating'] * 2) / 2
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df_review['review_text_word_count_cutoff'], bins=30, edgecolor='black')
plt.xlabel('Number of Words in Text', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title('Distribution of Word Count in Text', fontsize=16)

plt.subplot(1, 2, 2)
plt.hist(df_review['review_title_word_count'], bins=30, edgecolor='black')
plt.xlabel('Number of Words in Title', fontsize=16)
plt.ylabel('Frequency', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title('Distribution of Word Count in Title', fontsize=16)

plt.tight_layout()
plt.show()
```

```python
df_review['timestamp'] = pd.to_datetime(df_review['timestamp'])
df_review['year'] = df_review['timestamp'].dt.year
df_review['month'] = df_review['timestamp'].dt.month
df_review['day'] = df_review['timestamp'].dt.day
df_review['weekday'] = df_review['timestamp'].dt.weekday

df_review.drop(columns=['timestamp'], inplace=True)
```

```python
user_review_counts = df_review['user_id'].value_counts()

bins = [1, 2, float('inf')]
labels = ["1 review", "2+ reviews"]

user_review_counts_binned = pd.cut(user_review_counts, bins=bins, labels=labels, right=False)

user_review_proportions = user_review_counts_binned.value_counts(normalize=True) * 100

df_proportions = user_review_proportions.reset_index()
df_proportions.columns = ["Review Count", "User Percentage (%)"]

print(df_proportions)
```

```python
active_user= df_review['user_id'].value_counts() > 2

df_review['active_user'] = df_review['user_id'].map(active_user)
df_review.groupby('active_user')['verified_purchase'].count()

active_vs_helpful = df_review.groupby("active_user")["helpful_vote"].mean().reset_index()
active_vs_helpful
print("Average helpful vote for active vs non-active users:")
print(active_vs_helpful)

plt.figure(figsize=(8, 6))

plt.bar(active_vs_helpful["active_user"].astype(str), active_vs_helpful["helpful_vote"],
        color=['skyblue', 'salmon'])
plt.xlabel("Active User (False: Non-active, True: Active)")
plt.ylabel("Average Helpful Vote")
plt.title("Comparison of Average Helpful Vote for Active vs Non-active Users")
plt.tight_layout()
plt.show()
```

```python
grouped = df_review.groupby('rating').agg({
    'review_text_word_count_cutoff': 'mean',
    'review_title_word_count': 'mean'
}).reset_index()

grouped2 = df_meta.groupby('average_rating_threshold').apply(
    lambda x: np.average(x['meta_title_word_count'], weights=x['rating_number'])
).reset_index(name='meta_title_word_count')

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.barplot(x='rating', y='review_text_word_count_cutoff', data=grouped, ax=axes[0], color='#33A6B8')
axes[0].set_title("Review Text Word Count (Mean)", fontsize=16)
axes[0].set_xlabel("Rating", fontsize=16)
axes[0].set_ylabel("Mean Review Text Word Count", fontsize=16)
axes[0].tick_params(axis='both', labelsize=16)

sns.barplot(x='average_rating_threshold', y='meta_title_word_count', data=grouped2, ax=axes[1], color='#5DAC81')
axes[1].set_title("Meta Title Word Count (Mean)", fontsize=16)
axes[1].set_xlabel("Average Rating", fontsize=16)
axes[1].set_ylabel("Mean Meta Title Word Count", fontsize=16)
axes[1].tick_params(axis='both', labelsize=16)

plt.tight_layout()
plt.show()
```

```python
df_review['helpful_vote_binned'] = pd.cut(
    df_review['helpful_vote'],
    bins=[-np.inf, 0, 1, 2, 3, 4, 5, 10, np.inf],
    labels=['0', '1', '2', '3', '4', '5', '6-10', '11+']
)

grouped3 = df_review.groupby('helpful_vote_binned').agg({
    'review_text_word_count_cutoff': 'mean',
    'review_title_word_count': 'mean',
    'asin': 'count'
}).reset_index()

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sns.barplot(x='helpful_vote_binned', y='review_text_word_count_cutoff', data=grouped3, ax=axes[0], color='#33A6B8')
axes[0].set_title("Review Text Word Count (Mean)")
axes[0].set_xlabel("helpful_vote_binned")
axes[0].set_ylabel("Mean Review Text Word Count")

sns.barplot(x='helpful_vote_binned', y='review_title_word_count', data=grouped3, ax=axes[1], color='#5DAC81')
axes[1].set_title("Review Title Word Count (Mean)")
axes[1].set_xlabel("helpful_vote_binned")
axes[1].set_ylabel("Mean Review Title Word Count")

plt.show()
```

```python
df_review_filtered = df_review[(df_review['year'] >= 2012) & (df_review['year'] <= 2022)]
monthly_trend = df_review_filtered.groupby(['year', 'month']).size().unstack(level=0, fill_value=0)
first_group = [2012, 2013, 2014, 2015, 2016]
second_group = [2017, 2018, 2019]
third_group = [2020, 2021, 2022]

plt.figure(figsize=(8, 4))
for year in first_group:
    plt.plot(monthly_trend.index, monthly_trend[year], marker='o', linestyle='-', label=f'Year {year}')

plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.title('Monthly Review Trend (2012-2016)')
plt.xticks(ticks=range(1, 13), labels=[f'{m}' for m in range(1, 13)])
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(8, 4))
for year in second_group:
    plt.plot(monthly_trend.index, monthly_trend[year], marker='o', linestyle='-', label=f'Year {year}')

plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.title('Monthly Review Trend (2017-2019)')
plt.xticks(ticks=range(1, 13), labels=[f'{m}' for m in range(1, 13)])
plt.grid(True)
plt.legend()
plt.show()
#third
plt.figure(figsize=(8, 4))
for year in third_group:
    plt.plot(monthly_trend.index, monthly_trend[year], marker='o', linestyle='-', label=f'Year {year}')

plt.xlabel('Month')
plt.ylabel('Number of Reviews')
plt.title('Monthly Review Trend (2019-2022)')
plt.xticks(ticks=range(1, 13), labels=[f'{m}' for m in range(1, 13)])
plt.grid(True)
plt.legend()
plt.show()
```

```python
overall_monthly_trend = df_review_filtered.groupby('month').size()

plt.figure(figsize=(8, 4))
plt.plot(overall_monthly_trend.index, overall_monthly_trend.values, marker='o', linestyle='-', color='b', label='Overall Trend')
plt.xlabel('Month')
plt.ylabel('Total Number of Reviews')
plt.title('Overall Monthly Review Trend (2012-2022)')
plt.xticks(ticks=range(1, 13), labels=[f'{m}' for m in range(1, 13)])
plt.grid(True)
plt.legend()
plt.show()
```
